### 1.7.1 데이터 적재

```python
from sklearn.datasets import load_iris


iris_dataset = load_iris()
print(f'iris_datasets 키\n', iris_dataset.keys())
print(f'\n iris_datasets 형태\n', iris_dataset['data'].shape)
```

### 1.7.2 성과 측정: 훈련 데이터와 테스트 데이터

```python
from sklearn.model_selection import train_test_split


# sklearn.model_selection 제공 train_test_split(X dataset, Y dataset) : 자동으로 훈련, 테스트 데이터 구분
X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], iris_dataset['target'])
print(f'X_trian 크기: {X_train.shape}',f'\nX_test 크기: {X_test.shape}',
      f'\ny_train 크기: {y_train.shape}',f'\ny_test 크기: {y_test.shape}')
```

### 1.7.3 가장 먼저 할 일: 데이터 살펴보기

```python
import pandas as pd


# 산점도로 데이터 특성 확인
# 산점도는 두 개 특성만을 이용, 각각의 특성별 산점도 확인은 산점도 행렬 이용
# 특성의 수가 적은 경우에만 해당

# pd.plotting.scatter_matrix(데이터셋, ) : 산점도 행렬
i_df = pd.DataFrame(X_train, columns=iris_dataset.feature_names)
pd.plotting.scatter_matrix(i_df, c=y_train,
                          figsize=(15,15),
                          marker='o',
                          hist_kwds={'bins': 20})
```

### 첫번째 머신러닝 모델: k-최근접 이웃 알고리즘

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier


# KNeighborsClassifier(n_neighbors= k값) 
knn = KNeighborsClassifier(n_neighbors=1)

knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
print(f'모델 정확도: {np.mean(y_pred==y_test)}')
```
